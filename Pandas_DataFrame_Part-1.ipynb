```python
#if you gonna rename all the columns you can do an assignment and pass the new column names in a list
#schema_df.columns = ['Columns', 'SurveyQuestionText']
```

```python
#this is not commom and not a lot of use case for it
#instaed we will be changing column names to all upper/lower case, replacing space with underscore ect...
```

```python
#converting to upper case
#schema_df.reset_index(inplace = True)
schema_df.columns = [x.upper() for x in schema_df.columns]
schema_df.head()
```

```python
#let's change it to lower case
schema_df.columns = [x.lower() for x in schema_df.columns]
schema_df.head()
```

```python
#replacing space with undescore
schema_df.columns = schema_df.columns.str.replace(' ', '_')
schema_df.head()
#This won't change anything since there's no space in the column names
```

```python
#all the above methods applied to all the columns in the dataframe
#what if we want to chnage one column name?
#we can use '.rename' method
#we will be passing in old values and the new values in a dictionary to do so
```

```python
schema_df.rename(columns = {'column':'Column'}, inplace = True)
schema_df.head()
```

```python
schema_df.rename(columns = {'questiontext':'Questiontext'}, inplace = True)
schema_df.head()
```
