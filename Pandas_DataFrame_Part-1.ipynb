## Renaming columns all together or Selectively

If you gonna rename all the columns you can do an assignment and pass the new column names in a list

This is not commom and not a lot of use case for it. Instead we will be changing column names to all upper/lower case, replacing space with underscore ect...

```python
#converting to upper case
schema_df.columns = [x.upper() for x in schema_df.columns]
schema_df.head()
```

```python
#let's change it to lower case
schema_df.columns = [x.lower() for x in schema_df.columns]
schema_df.head()
```

```python
#replacing space with undescore
schema_df.columns = schema_df.columns.str.replace(' ', '_')
schema_df.head()
#This won't change anything since there's no space in the column names
```

All the above methods applied to all the columns in the dataframe. But, what if we want to chnage one column name?. We can use '.rename' method. We will be passing in old values and the new values in a dictionary to do so

```python
schema_df.rename(columns = {'column':'Column'}, inplace = True)
schema_df.head()
```

```python
schema_df.rename(columns = {'questiontext':'Questiontext'}, inplace = True)
schema_df.head()
```

### Advanced Methods to change the values of the columns

#### There are 4 advance ways to do this:
1. apply
2. map
3. applymap
4. replace

These 4 methods are used in different scenarios. We'll go over them below one by one.

### 1- Apply

We can call a function that can applied to all the values. It can work on DataFrame or Series object.

#### How Apply works in a Series?

When we apply .apply in a series it applies the function to every value in the series.

```python
#let's apply .len function to find out all the lengths of Columns in the schema_df DataFrame.
schema_df['Column'].apply(len)
```

```python
#let's update the Column names to all UPPER case values
#we'll create a function to do so and apply it to the Column
#we could have just used simple .upper function instead, but for the sake of this example we'll use .apply
```

```python
def updating_email(Column):
    return Column.upper()
```

```python
#lets apply it to the Column now
schema_df['Column'].apply(updating_email)
```

```python
schema_df.head()
```

```python
#we can also use a lambda function
#lets make the Column lower case
schema_df['Column'].apply(lambda x: x.lower())
```

#### How Apply works on DataFrame?

When we apply .apply on dataFrame it runs a function on each row or column. Contrary to series when we use apply on a DataFrame, it applies to every series in the DataFrame instead of every value.

```python
schema_df.apply(len)
```

```python
#it denotes how many values are there in each column in the dataframe
```

 we can do the same by manually checking each column using len function

```python
len(schema_df['Column'])
```

```python
len(schema_df['Questiontext'])
```

```python
#let's count by values the rows
#it'll count how many values are there in each row
schema_df.apply(len, axis='columns')
```

```python
#we can apply the min function from the .series to get the minimum value
#ideally this would be useful on numerical values to use with numpy
schema_df.apply(pd.Series.min)
```

```python
#we can do the same by using lambda function
schema_df.apply(lambda x: x.min())
```

#### How does Applymap works?

How do we apply these function to every value in a DataFrame? We can use .applymap for that.

```python
#This will return the lenth of all the value in this DataFrame
schema_df.applymap(len)
```

```python
#Let's say we wants to make every value in df a lower case
#since we have  strings in the df the string function .lower would work here
schema_df.applymap(str.lower)
```

#### How does map method work?

Map method only work on series objects. IT's used to substituting each value with another value. 

```python
#let's change some column names
schema_df['Column'].map({'trans':'transgnder','surveyease':'surveyeasing'})
```


values we didn't substitute will change to NaN values. Well, if we didn't want that to happen we can use .replace instead of .map method.

#### How does .replace method works?

```python
#Let's do the same we did with map using replace.
schema_df['Column'].replace({'trans':'transgnder','surveyease':'surveyeasing'})
```

```python
#this gave us exactly what we wanted. 
#In addition it didn't chnage the other values to NaN values like .map did.
```
